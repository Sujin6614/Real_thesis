```python
#Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Loading the dataset for phisisng detection system: Lnk: https://www.kaggle.com/datasets/shashwatwork/phishing-dataset-for-machine-learning/data

data = pd.read_csv('Phishing_Legitimate_full.csv')
```

```python
data.drop("id", axis=1, inplace=True)
```

```python
data.head()
```

```python
data['CLASS_LABEL'].hist()
```

```python
#Splitting data
X = data.drop('CLASS_LABEL', axis=1)
y = data['CLASS_LABEL']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.shape
```

```python
#Traning Random Forest Algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
preds=rfc.predict(X_test)
```

```python
#Model evaluation metrics
sum(preds==y_test)/len(preds)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
accuracy_score(y_test, preds)
```

```python
precision_score(y_test, preds)
```

```python
recall_score(y_test, preds)
```
